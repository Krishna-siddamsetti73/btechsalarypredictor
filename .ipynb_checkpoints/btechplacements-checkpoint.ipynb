```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("manishkc06/engineering-graduate-salary-prediction")

print("Path to dataset files:", path)
```

```python
import pandas as pd
df = pd.read_csv(path + "/Engineering_graduate_salary.csv")
print(df.head())
print(df.describe())
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df=df.drop_duplicates()
degr = df['Degree'].value_counts()
degree=df['Degree'].unique()
specialization = df['Specialization'].value_counts()
#print(degree , specialization, degr) 
df=df[df['Degree'] == 'B.Tech/B.E.']
#print(df['Degree'].unique())
#print(len(df['Specialization'].unique()))
#print(df['Specialization'].value_counts())
specializations = specialization[specialization > 10]
print(specializations)
df.Specialization = df.Specialization.apply(lambda x: x if x in specializations  else 'Other')
print(df['Specialization'].value_counts())
df = df[(df['10board'] == 'cbse') & (df['12board'] == 'cbse')]
df.head()
df = df.replace(-1, np.nan)
cols_with_nan = [column for column in df.columns if df.isna().sum()[column] > 0]
for column in cols_with_nan:
    df[column] = df[column].fillna(df[column].mean())
df = df.drop(['ID', 'DOB', '10board'], axis=1)
```

```python
x=df['Specialization']
y=df['collegeGPA']
z=df['Salary']
plt.figure(figsize=(10,6))
plt.scatter(x,z,c=y,cmap='viridis',marker='4')
plt.colorbar(label='collegegpa')
plt.xlabel('Specialization')
plt.ylabel('salary')
plt.title('Specialization vs collegeGPA vs Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
das = df['Salary'].groupby(df['Specialization']).mean()
dd = []
s = df.groupby('Specialization')['Salary'].apply(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)

for spec, j, i in zip(df['Specialization'].unique(), s, das):
    dd.append(f"{spec}:mode:{j}:mean:{int(i)}")

plt.pie(das, labels=dd, autopct='%1.1f%%')
plt.title('Specialization vs Salary')
plt.show()
```
